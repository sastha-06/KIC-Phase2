```python
plt.figure()
plt.scatter(df['Log GDP per capita'], df['HappinessScore'], color='red')
plt.xlabel('Log GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Scatter Plot: GDP vs Happiness')
plt.show()
```

```python

plt.figure()
sns.scatterplot(x=df['Healthy life expectancy at birth'], y=df['HappinessScore'],color='green')
plt.xlabel('Healthy life expectancy')
plt.ylabel('Happiness Score')
plt.title('Scatter Plot: GDP vs Happiness')
plt.show()
```

```python
df_sorted = df.sort_values('Rank')
plt.figure()
sns.lineplot(x='Rank', y='HappinessScore',data=df_sorted)
plt.xlabel('Rank')
plt.ylabel('Happiness Score')
plt.title('Line Plot: Rank vs Happiness Score')
plt.show()
```

```python
top10 = df.sort_values('HappinessScore', ascending=False).head(10)

plt.figure()
plt.bar(top10['Country'], top10['HappinessScore'])
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Top 10 Happiest Countries')
plt.show()
```

```python
region_avg = df.groupby('Region')['HappinessScore'].mean()
plt.figure()
region_avg.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.title('Average Happiness by Region')
plt.show()
```

```python
plt.figure()
sns.histplot(df['HappinessScore'], bins=10,kde=True)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Region',y='HappinessScore',data=df)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.title('Happiness Score by Region')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x=df['Freedom to make life choices'], y=df['HappinessScore'],color='yellow')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
plt.title('Freedom vs Happiness')
plt.show()
```

```python
plt.figure()
sns.lineplot(df['Positive affect'], label='Positive Affect')
sns.lineplot(df['Negative affect'], label='Negative Affect')
plt.xlabel('Index')
plt.ylabel('Affect Score')
plt.title('Positive vs Negative Affect')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='Region',y='HappinessScore',data=df)
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.title('Swarm Plot: Region vs Happiness Score')
plt.show()
```

```python
num_df = df.select_dtypes(include='number')

plt.figure(figsize=(12,8))
sns.heatmap(num_df.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap of Happiness Dataset')
plt.show()
```
