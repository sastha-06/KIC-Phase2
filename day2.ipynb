```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

params = [(10, 0.5), (10, 0.2), (20, 0.5), (20, 0.7)]

plt.figure(figsize=(10, 6))

for n, p in params:
    x = np.arange(0, n+1)
    y = binom.pmf(x, n, p)

    plt.plot(x, y, marker='o', label=f'n={n}, p={p}')
plt.title("Binomial Distribution for Different n and p")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()
plt.show()

for n, p in params:
    x = np.arange(0, n+1)
    y = binom.pmf(x, n, p)
    plt.bar(x, y,label=f'n={n}, p={p}')
plt.title("Binomial Distribution for Different n and p")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
