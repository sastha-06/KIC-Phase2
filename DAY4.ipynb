```python
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind

df = sns.load_dataset("titanic")

survivors = df[df["survived"] == 1]["age"].dropna()
non_survivors = df[df["survived"] == 0]["age"].dropna()
t_stat, p_value = ttest_ind(survivors, non_survivors, equal_var=False)
print("Average Age of Survivors:", round(survivors.mean(), 2))
print("Average Age of Non-Survivors:", round(non_survivors.mean(), 2))
print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 6))
```
